```python
import pandas as pd
schools_df=pd.read_csv(r"D:\MONASH Bootcamp\monu-mel-data-pt-08-2021-u-c\02-Homework\04-Pandas\Instructions\PyCitySchools\Resources\schools_complete.csv")
students_df=pd.read_csv(r"D:\MONASH Bootcamp\monu-mel-data-pt-08-2021-u-c\02-Homework\04-Pandas\Instructions\PyCitySchools\Resources\students_complete.csv")
schools_df=schools_df.dropna(how='any')

students_df=students_df.dropna(how='any')
students_df.count(),schools_df.count()
#Anlyze all the columns
schools_df.columns,students_df.columns

schools_df=schools_df.rename(columns={"School ID":"school_id"})
schools_df.head()

#Create a high level snapshot (in table form) of the district's key metrics, 
##rename student id
students_df=students_df.rename(columns={"Student ID":"student_id"})
students_df.head()
students_df["student_id"].value_counts()
students_df#unique values and all non numeric data also
merged_df=pd.merge(schools_df,students_df,on=["school_name"])
merged_df
#Clean Data
a=merged_df.count()
merged_df=merged_df.dropna(how='any')
b=merged_df.count()
a,b
```

```python

#-------------------------------District Summary-------------------------------
#Total Schools
#Total Students
Total_Schools=merged_df["school_id"].nunique()
Total_Students=merged_df["student_id"].nunique()
#Total Budget
unique_Budget=merged_df["budget"].unique()
Total_Budget=unique_Budget.sum()

#Average Math Score
Average_Math_Score=round(merged_df["math_score"].mean(),)
#Average Reading Score
Average_Reading_Score=round(merged_df["reading_score"].mean(),)
#% Passing Math (The percentage of students that passed math.)
Pass_Math=merged_df.loc[merged_df["math_score"]>70,"math_score"].count()
total_scores=merged_df["math_score"].count()
Percent_Pass_Math=round(100*Pass_Math/total_scores,2)
# % Passing Reading (The percentage of students that passed reading.)
Pass_Reading=merged_df.loc[merged_df["reading_score"]>70,"reading_score"].count()
total_scores=merged_df["reading_score"].count()
Percent_Pass_Reading=round(100*Pass_Reading/total_scores,2)
#% Overall Passing (The percentage of students that passed math and reading.)
Pass_Both=merged_df.loc[(merged_df["reading_score"]>70)&(merged_df["math_score"]>70),"math_score"].count()
total_scores=merged_df["reading_score"].count()
Percent_Pass_Overall=round(100*Pass_Both/total_scores,2)
 
#Create a high level snapshot (in table form) of the district's key metrics, including:
snapshot_df=pd.DataFrame({"Average_Math_Score":[Average_Math_Score],
               "Average_Reading_Score":Average_Reading_Score,
               "Percent_Pass_Math":Percent_Pass_Math,
              "Percent_Pass_Reading":Percent_Pass_Reading,
               "Percent_Pass_Overall":Percent_Pass_Overall})
snapshot_df.head()
```

```python
#-------------------------------School Summary-------------------------------
#School Name
School_Name=merged_df["school_name"].unique()
#School Type
school_type = schools_df.set_index(["school_name"])["type"]
#Total Students
students_per_school=merged_df.groupby("school_name")["student_id"].count()
#Total School Budget
budget_per_school = merged_df.groupby(["school_name"])["budget"].mean()
budget_per_school.head()
len(budget_per_school)
#Per Student Budget
psb=budget_per_school/students_per_school
#Average Math Score
Average_Math_Score = round(merged_df.groupby(["school_name"])["math_score"].mean())
#Average Reading Score
Average_Read_Score = round(merged_df.groupby(["school_name"])["reading_score"].mean())
Average_Read_Score
PASS_PER_SCHOOL=merged_df.loc[merged_df["math_score"]>70,["student_name","school_name"]]
students_passed_in_schools=PASS_PER_SCHOOL.groupby("school_name")["student_name"].count()
#% Passing Math (The percentage of students that passed math.)
Percent_Pass_Math=(students_passed_in_schools/students_per_school)*100
##% Passing Reading (The percentage of students that passed reading.)
PASS_PER_SCHOOL=merged_df.loc[merged_df["reading_score"]>70,["student_name","school_name"]]
students_passed_reading_in_schools=PASS_PER_SCHOOL.groupby("school_name")["student_name"].count()
#% Passing Math (The percentage of students that passed math.)
Percent_Pass_Reading=(students_passed_reading_in_schools/students_per_school)*100
#% Overall Passing (The percentage of students that passed math and reading.)
PASS_PER_SCHOOL_both=merged_df.loc[(merged_df["reading_score"]>70)&(merged_df["math_score"]>70),["reading_score","math_score","school_name","student_name"]]
students_passed_both_in_schools=PASS_PER_SCHOOL_both.groupby("school_name")["student_name"].count()
#% Passing Math (The percentage of students that passed math.)
Percent_Pass=(students_passed_both_in_schools/students_per_school)*100
print(Percent_Pass)
```

```python
#-------------------------------School Summary-------------------------------
snapshot_df2=pd.DataFrame({
                         "School Type":school_type,
                                  "Total Students":students_per_school,
                                 "Total School Budget":budget_per_school,
                                  "Budget per student":psb,
                                 "Average Math Score":Average_Math_Score,
                                 "Average Reading Score":Average_Read_Score,
                                  "% Passing Math":Percent_Pass_Math,
                                  "% Passing Reading":students_passed_reading_in_schools,
                                  "Overall Passing Rate":Percent_Pass})
snapshot_df2.head()



    
```

```python
#------------------------------Top Performing Schools (By % Overall Passing)--------------------------------
Top5Schools=snapshot_df2.sort_values(["Overall Passing Rate"],ascending=False).head()
Top5Schools
#------------------------------Weakest 5 Performing Schools (By % Overall Passing)--------------------------------
Bottom5Schools=snapshot_df2.sort_values(["Overall Passing Rate"],ascending=False).tail(5)
Bottom5Schools.sort_values(["Overall Passing Rate"],ascending=True)#Lowest at top
#-------------------------------avg_math_scores_grade by School-------------------------------
avg_math_scores_grade = round(merged_df.groupby(["grade"]).mean()["math_score"],2)
avg_math_scores_grade
#-------------------------------avg_reading_score_grade by School -------------------------------
avg_reading_score_grade = round(merged_df.groupby(["grade"]).mean()["reading_score"],2)
avg_reading_score_grade
```
